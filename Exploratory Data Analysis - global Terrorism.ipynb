```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Reading the data

```python
df = pd.read_csv('D:/GRIP/globalterrorism.csv', encoding = "ISO-8859-1", engine='python')
```

```python
df2 = df.copy()                           #copying the dataset into new dataframe
```

```python
df2.head()
```

```python
df2.dropna(axis=1, how='all')                  #dropping the columns which have null values
```

### Renaming the columns, to read and understand easily

```python
df2.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'state',
                       'region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed',
                       'nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                       'weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

#### Considering only usefull columns

```python
df2=df2[['Year','Month','Day','Country','state','Region','city','latitude','longitude','AttackType','Killed',
               'Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]
df2['Wounded'] = df2['Wounded'].fillna(0).astype(int)
df2['Killed'] = df2['Killed'].fillna(0).astype(int)                            #filling the null value with zeros
```

```python
df2['casualities']=df2['Killed']+df2['Wounded']            #adding a new column 'casuality'
```

```python
df2.head()
```

```python
df2.isnull().sum()
```

### Basic analysis of the data

```python
print("The country with highest terror attack:",df2['Country'].value_counts().index[0])
print("Regions with highest terrorist attacks: ",df2['Region'].value_counts().index[0])
print("Highest people killed in an attack",df2['Killed'].max(),"that too place in",df2.loc[df2['Killed'].idxmax()].Country)
print("City with the highest number of attacks:",df2['city'].value_counts().index[1])
print("Year with the most attacks:",df2['Year'].value_counts().idxmax())
print("Month with the most attacks:",df2['Month'].value_counts().idxmax())
print("Group with the most attacks:",df2['Group'].value_counts().index[1])
print("Most Attack Types:",df2['AttackType'].value_counts().idxmax())
```

## Comparing the number of Terrorist Attacks and the number of Dead/Injured people from 1970 to 2017

```python
#Filter Data to choose Number Of Dead/Injured People In Terrorist Attacks From 1970 to 2017
people_damage = df2[["Year","casualities"]].groupby('Year').sum()
list_year =  df2["Year"].unique().tolist()

#draw bar chart
fig, ax1 = plt.subplots(figsize = (20,6))
ax1.bar(people_damage.index, [i[0] for i in people_damage.values], color= '#0063B1' )

ax1.set_xticklabels(np.arange(1970, 2018, step=1), rotation=90)
ax1.set_ylabel('Number Of Dead/Injured People', size = 12)
ax1.set_xlabel('Year', size = 12)
ax1.set_title('Number of Terrorist Attacks vs Number of Dead/Injured people From 1970 to 2017', fontsize= 15, pad= 10, weight ='bold', 
                    color = sns.cubehelix_palette(8, start=.5, rot=-.75)[-3])
ax2 = ax1.twinx()

#Filter & get a number of attacked in the world from 1970 to 2017 
number_attack = []
for year in list_year:
    number_attack.append(len(df2[df2['Year'] == year][["Year"]]))
number_attack.insert(23, 0)

#draw plot chart
ax2.set_ylabel('Number Of Terrorist Attacks', size = 12,rotation=-90)
ax2.plot(range(1970, 2018), number_attack, 'r--o', mfc='k', label='Number Of Terrorist Attacks')

plt.xticks(np.arange(1970, 2018, step=1))
plt.legend(loc='upper left')
plt.show()
```

### The below plot shows the most affected countries

```python
plt.subplots(figsize=(18,6))
ax=sns.barplot(df2['Country'].value_counts()[:15].index,df2['Country'].value_counts()[:15].values,palette='magma')

for rect in ax.patches:
   
    height = rect.get_height()
    width = rect.get_width()
    x = rect.get_x()
    y = rect.get_y()
    
    label_text = f'{height}'  
    
    label_x = x + width / 2
    label_y = y + height / 2

    if height > 0:
        ax.text(label_x, label_y, label_text, ha='center', va='center', color='white', fontsize=8)

plt.title('Top Affected Countries', fontsize=20)
plt.show()
```

## From the above plot we can see that the Iraq is the most affected country by terrorism

## Plotting the graph of four most affected countries

```python
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
```

```python
hover_text = []
for index, row in terror_df.iterrows():
    hover_text.append(('City: {city}<br>'+'Group: {group}<br>'+'casualities: {casualities}<br>'+
                       'Year: {Year}').format(city=row['city'],
                                            group=row['Group'],
                                            casualities=row['casualities'],
                                            Year=row['Year']))
terror_df['text'] = hover_text
```

```python
trace0 = go.Scatter(
    x=terror_df['Year'][terror_df['Country'] == 'Iraq'],
    y=terror_df['casualities'][terror_df['Country'] == 'Iraq'],
    mode='markers',
    name='Iraq',
    text=terror_df['text'][terror_df['Country'] == 'Iraq'],
    marker=dict(
        symbol='circle',
        sizemode='area',
        size=terror_df['casualities'][terror_df['Country'] == 'Iraq'],
        line=dict(width=2),)
)

trace1 = go.Scatter(
    x=terror_df['Year'][terror_df['Country'] == 'Pakistan'],
    y=terror_df['casualities'][terror_df['Country'] == 'Pakistan'],
    mode='markers',
    name='Pakistan',
    text=terror_df['text'][terror_df['Country'] == 'Pakistan'],
    marker=dict(
        symbol='circle',
        sizemode='area',
        size=terror_df['casualities'][terror_df['Country'] == 'Pakistan'],
        line=dict(width=2),)
)

trace2 = go.Scatter(
    x=terror_df['Year'][terror_df['Country'] == 'Afghanistan'],
    y=terror_df['casualities'][terror_df['Country'] == 'Afghanistan'],
    mode='markers',
    name='Afghanistan',
    text=terror_df['text'][terror_df['Country'] == 'Afghanistan'],
    marker=dict(
        symbol='circle',
        sizemode='area',
        size=terror_df['casualities'][terror_df['Country'] == 'Afghanistan'],
        line=dict(width=2),)
)

trace3 = go.Scatter(
    x=terror_df['Year'][terror_df['Country'] == 'India'],
    y=terror_df['casualities'][terror_df['Country'] == 'India'],
    mode='markers',
    name='India',
    text=terror_df['text'][terror_df['Country'] == 'India'],
    marker=dict(
        symbol='circle',
        sizemode='area',
        size=terror_df['casualities'][terror_df['Country'] == 'India'],
        line=dict(width=2),)
)
```

```python
data = [trace0, trace1, trace2, trace3]
layout = go.Layout(
         title = 'The Four most affected countries by terrorism',
         xaxis = dict(
             title = 'Year',
             range = [1976,2016],
             tickmode = 'auto',
             nticks = 30,
             showline = True,
             showgrid = False,color='black'
             ),
         yaxis = dict(
             title = 'Casualities',
             type = 'log',
             range = [1.8,3.6],
             tickmode = 'auto',
             nticks = 40,
             showline = True,
             showgrid = False,color='black'),
    paper_bgcolor='rgb(130, 225, 238)',
    plot_bgcolor='rgb(130, 225, 238)')
```

```python
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='Terrorism Bubble')
```

## The below map shows the parts of the countries that are attacked by the top 8 terrorist organisations

```python
import folium
from folium.plugins import MarkerCluster
```

```python
terror_group = df2.dropna(subset=['latitude','longitude'])
```

```python
terror_group = terror_group.drop_duplicates(subset=['Country','Group'])
```

```python
terrorist_groups = df2.Group.value_counts()[1:8].index.tolist()
```

```python
terror_group = terror_group.loc[terror_group.Group.isin(terrorist_groups)]
```

```python
terror_group.Group.unique()
```

```python
m1 = folium.Map(location=[20, 0], tiles="Stamenterrain", zoom_start=2)
marker_cluster = MarkerCluster(
    name='clustered icons',overlay=True,control=False,icon_create_function=None)
for i in range(0,len(terror_group)):
    marker=folium.Marker([terror_group.iloc[i]['latitude'],terror_group.iloc[i]['longitude']]) 
    popup='Group:{}<br>Country:{}'.format(terror_group.iloc[i]['Group'],
                                          terror_group.iloc[i]['Country'])
    folium.Popup(popup).add_to(marker)
    marker_cluster.add_child(marker)
marker_cluster.add_to(m1)
folium.TileLayer('openstreetmap').add_to(m1)
folium.TileLayer('Mapbox Bright').add_to(m1)
folium.TileLayer('cartodbdark_matter').add_to(m1)
folium.TileLayer('stamentoner').add_to(m1)
folium.LayerControl().add_to(m1)
```
